```python
# ===============================
# 📦 IMPORT LIBRARIES
# ===============================
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

sns.set(style="whitegrid")

# ===============================
# 🔗 CONNECT TO SQL SERVER
# ===============================
conn = pyodbc.connect(
    "Driver={ODBC Driver 17 for SQL Server};"
    r"Server=DESKTOP-OO5HFAI\SQLEXPRESS;"
    "Database=Data_Spark;"
    "Trusted_Connection=yes;"
)

# ===============================
# 📥 LOAD TABLES
# ===============================
customers = pd.read_sql("SELECT * FROM Customers", conn)
sales = pd.read_sql("SELECT * FROM Sales", conn)
products = pd.read_sql("SELECT * FROM Products", conn)
stores = pd.read_sql("SELECT * FROM Stores", conn)
exchange = pd.read_sql("SELECT * FROM Exchange_Rates", conn)

print("Customers:", customers.shape)
print("Sales:", sales.shape)
print("Products:", products.shape)
print("Stores:", stores.shape)
print("Exchange Rates:", exchange.shape)

# ===============================
# 🧑 CUSTOMER ANALYSIS
# ===============================
today = datetime.today()
customers["Age"] = pd.to_datetime(customers["Birthday"], errors="coerce").apply(
    lambda x: today.year - x.year if pd.notnull(x) else None
)
bins = [0, 20, 30, 40, 50, 60, 100]
labels = ["<20", "21-30", "31-40", "41-50", "51-60", "60+"]
customers["AgeGroup"] = pd.cut(customers["Age"], bins=bins, labels=labels, right=False)

# Gender Distribution
gender_dist = customers["Gender"].value_counts(normalize=True) * 100
plt.pie(gender_dist, labels=gender_dist.index, autopct='%1.1f%%', startangle=90)
plt.title("Gender Distribution")
plt.show()

# Age Distribution
plt.hist(customers["Age"].dropna(), bins=20, color="skyblue", edgecolor="black")
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# Age Groups
age_group = customers["AgeGroup"].value_counts().sort_index()
sns.barplot(x=age_group.index, y=age_group.values, palette="viridis")
plt.title("Customers by Age Group")
plt.show()

# Customers per Country (Top 10)
country_counts = customers["Country"].value_counts().head(10)
sns.barplot(x=country_counts.values, y=country_counts.index, palette="magma")
plt.title("Top 10 Countries by Customers")
plt.xlabel("Customers")
plt.show()

# ===============================
# 💰 SALES & PRODUCTS ANALYSIS
# ===============================

# Merge Sales + Products for Revenue
sales_products = sales.merge(products, on="ProductKey", how="left")
sales_products["Revenue"] = sales_products["Quantity"] * sales_products["Unit Price USD"]

# Total Orders per Customer
orders_per_customer = sales.groupby("CustomerKey")["Order Number"].nunique()
cust_orders = customers.merge(orders_per_customer, left_on="CustomerKey", right_index=True, how="left")
cust_orders.rename(columns={"Order Number": "TotalOrders"}, inplace=True)

# Average Order Size
avg_order_size = sales.groupby("CustomerKey")["Quantity"].mean()
cust_orders = cust_orders.merge(avg_order_size, left_on="CustomerKey", right_index=True, how="left")
cust_orders.rename(columns={"Quantity": "AvgOrderSize"}, inplace=True)

# Customer Lifetime Value (Total Revenue)
cust_ltv = sales_products.groupby("CustomerKey")["Revenue"].sum().reset_index()
cust_orders = cust_orders.merge(cust_ltv, on="CustomerKey", how="left")
cust_orders.rename(columns={"Revenue": "TotalRevenue"}, inplace=True)

# Top 10 Customers by Revenue
top_customers = cust_orders.sort_values("TotalRevenue", ascending=False).head(10)
sns.barplot(x="TotalRevenue", y="Name", data=top_customers, palette="coolwarm")
plt.title("Top 10 Customers by Revenue")
plt.show()

# Revenue by Product Category
sales_category = sales_products.groupby("Category")["Revenue"].sum().sort_values(ascending=False)
sns.barplot(x=sales_category.values, y=sales_category.index, palette="viridis")
plt.title("Revenue by Product Category")
plt.xlabel("Revenue")
plt.show()

# Top 10 Products by Revenue
top_products = sales_products.groupby("Product Name")["Revenue"].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_products.values, y=top_products.index, palette="plasma")
plt.title("Top 10 Products by Revenue")
plt.xlabel("Revenue")
plt.show()

# Price vs Revenue Correlation
price_revenue = sales_products.groupby("ProductKey").agg({
    "Unit Price USD": "mean",
    "Revenue": "sum"
}).reset_index()
sns.scatterplot(x="Unit Price USD", y="Revenue", data=price_revenue)
plt.title("Price vs Revenue per Product")
plt.xlabel("Unit Price USD")
plt.ylabel("Revenue")
plt.show()

# ===============================
# 🏬 STORES ANALYSIS
# ===============================

# Stores by Country (Top 10)
store_country = stores["Country"].value_counts().head(10)
sns.barplot(x=store_country.values, y=store_country.index, palette="coolwarm")
plt.title("Top 10 Countries by Store Count")
plt.xlabel("Stores")
plt.show()

# Store Size Distribution
sns.histplot(stores["Square Meters"].dropna(), bins=20, kde=True, color="orange")
plt.title("Distribution of Store Sizes")
plt.xlabel("Square Meters")
plt.show()

# Store Open Date Trend
stores["Open Date"] = pd.to_datetime(stores["Open Date"], errors="coerce")
open_trend = stores["Open Date"].dt.year.value_counts().sort_index()
sns.lineplot(x=open_trend.index, y=open_trend.values, marker="o")
plt.title("Stores Opened Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Stores")
plt.show()

# Average Store Size per Country (Top 10)
avg_size_country = stores.groupby("Country")["Square Meters"].mean().sort_values(ascending=False).head(10)
sns.barplot(x=avg_size_country.values, y=avg_size_country.index, palette="Blues_r")
plt.title("Average Store Size per Country")
plt.xlabel("Square Meters")
plt.show()

# Revenue per Store (Top 10)
sales_stores = sales_products.groupby("StoreKey")["Revenue"].sum().reset_index()
store_revenue = stores.merge(sales_stores, on="StoreKey", how="left")
sns.barplot(
    x="Revenue",
    y="StoreKey",  # using StoreKey as label
    data=store_revenue.sort_values("Revenue", ascending=False).head(10),
    palette="viridis"
)
plt.title("Top 10 Stores by Revenue")
plt.show()

# ===============================
# 💱 EXCHANGE RATES ANALYSIS
# ===============================
exchange["Date"] = pd.to_datetime(exchange["Date"], errors="coerce")

# Records per Currency
currency_count = exchange["Currency"].value_counts()
sns.barplot(x=currency_count.values, y=currency_count.index, palette="plasma")
plt.title("Exchange Rate Records by Currency")
plt.xlabel("Records Count")
plt.show()

# Trend of EUR Exchange Rate
eur_trend = exchange[exchange["Currency"] == "EUR"].set_index("Date")["Exchange"]
eur_trend.plot(figsize=(10,4), color="green")
plt.title("EUR Exchange Rate Trend")
plt.xlabel("Date")
plt.ylabel("Exchange Rate")
plt.show()

# Exchange Rate Volatility by Currency
currency_vol = exchange.groupby("Currency")["Exchange"].std().sort_values(ascending=False)
sns.barplot(x=currency_vol.values, y=currency_vol.index, palette="inferno")
plt.title("Exchange Rate Volatility (Std Dev)")
plt.xlabel("Standard Deviation")
plt.show()

# Max & Min Exchange Rate per Currency
currency_maxmin = exchange.groupby("Currency")["Exchange"].agg(["min","max"]).sort_values("max", ascending=False)
print("Max & Min Exchange Rate per Currency:\n", currency_maxmin)
```
